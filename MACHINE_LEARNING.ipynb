**ONE HOT ENCODER**

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
df=pd.DataFrame({"IPL":["RCB","MI","LSG","PBKS","DC","RR","SRH","GT","CSK",]})
data=ohe.fit_transform(df[['IPL']])
names=ohe.get_feature_names_out(['IPL'])
df1=pd.DataFrame(data,columns=names)
df1
```

```python
df=pd.DataFrame({"Fruits":["Apple","Orange","Grapes","Apple","Grapes"]})
data=ohe.fit_transform(df[['Fruits']])
names=ohe.get_feature_names_out(['Fruits'])
df1=pd.DataFrame(data,columns=names)
df1
```

**ORDINAL ENCODING**

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Dist','FC','SC','TC','Fail']])
df=pd.DataFrame({"Ranks":["Dist","FC","SC","Fail","Dist",'Fail','FC']})
df['Ranks']=oe.fit_transform(df[['Ranks']])
df
```

**FEATURE SCALING**

* **NORMALISATION / L2 NORM / EUCLIDEAN NORM**

```python
import pandas as pd
from sklearn.preprocessing import Normalizer
nor=Normalizer()
df=pd.DataFrame({"Weight":[65,80,75],"Height":[170,130,150]})
data=nor.fit_transform(df)
df1=pd.DataFrame(data,columns=df.columns)
df1
```

```python
import pandas as pd
from sklearn.preprocessing import Normalizer
nor=Normalizer()
df=pd.DataFrame({"Exp":[1,2,3,4,5],"Sal":[35000,45000,50000,60000,75000]})
data=nor.fit_transform(df[['Exp','Sal']])
df1=pd.DataFrame(data,columns=df.columns)
df1
```

* **Standard Scalar / Z2-Score**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df=pd.DataFrame({"Weight":[65,80,75],"Height":[170,130,110]})
data=sc.fit_transform(df)
df1=pd.DataFrame(data,columns=df.columns)
df1
```

MANUAL CALCULATION


```python
import numpy as np
h=[72,55,50]
mh=np.mean(h)

print('height avg: ',mh)
#sum(xi-u)2
for i in h:
  ms=(i-mh)**2
  print(i,ms)
```

```python
total=(169+16+81)
sd=np.sqrt(total/3)
print(sd)
```

```python
h=[72,55,50]
for i in h:
  print((i-mh)/sd)
```
