**ONE HOT ENCODER**

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
df=pd.DataFrame({"IPL":["RCB","MI","LSG","PBKS","DC","RR","SRH","GT","CSK",]})
data=ohe.fit_transform(df[['IPL']])
names=ohe.get_feature_names_out(['IPL'])
df1=pd.DataFrame(data,columns=names)
df1
```

```python
df=pd.DataFrame({"Fruits":["Apple","Orange","Grapes","Apple","Grapes"]})
data=ohe.fit_transform(df[['Fruits']])
names=ohe.get_feature_names_out(['Fruits'])
df1=pd.DataFrame(data,columns=names)
df1
```

```python
data=pd.DataFrame({"Bikes":["Classic 350","Scrambler","Duke 250","Scrambler","Classic 350"]})
df=ohe.fit_transform(data[["Bikes"]])
names=ohe.get_feature_names_out(["Bikes"])
df1=pd.DataFrame(df,columns=names)
df1
```
